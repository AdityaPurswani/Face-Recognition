```python
class CustomKNN:
  #constructor
  def __init__(self, n_neighbors=5):
    self.n_neighbors = n_neighbors

  # training function
  def fit(self, X, y):
    self._X = X.astype(np.int64)
    self._y = y

  # predict point
  # given a single point, tell me which class it belongs to
  def predict_point(self, point):
    # storing the dis of given 'point' from each point in training data
    list_dist = []

    #these point are from the training database 
    for x_point, y_point in zip(self._X, self._y):
      dist_point = ((point - x_point)**2).sum()
      list_dist.append([dist_point, y_point])

    # sorting the list according to the distance
    sorted_dist = sorted(list_dist)
    top_k_nearest = sorted_dist[:self.n_neighbors]

    # taking the count
    items, count = np.unique(np.array(top_k_nearest)[:, 1], return_counts=True)
    ans = items[np.argmax(count)]
    return ans

  #predict
  # give an answer for each number in the array
  def predict(self, X):
    results = []
    for point in X:
      results.append(self.predict_point(point))
    return np.array(results)

  # Score to measure accuracy
  def score(self, X, y):
    return sum(self.predict(X) == y)/ len(y)
```

```python
import numpy as np

data = np.load("face_data.npy")
model = CustomKNN()

X = data[:, 1:].astype(int)
y = data[:, 0]

model.fit(X, y)
```

```python
import cv2
import numpy as np
import os
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = detector.detectMultiScale(frame)

    for face in faces:
        x, y, w, h = face 
        cut = frame[y:y+h, x:x+w]

        fix = cv2.resize(cut, (100, 100))
        gray = cv2.cvtColor(fix, cv2.COLOR_BGR2GRAY)

        out = model.predict(np.array([gray.flatten()]))
        
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 255, 255), 2)
        cv2.putText(frame, str(out[0]), (x, y-20), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 4)
        print(out)
        cv2.imshow("My Face", gray)
    cv2.imshow("My Frame", frame)
    
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
